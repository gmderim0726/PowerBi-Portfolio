```python
import pandas as pd
import os
from sqlalchemy import create_engine

caminho = r"E:\Adress"
usuario = "USER"
senha="PASSWORD"
host="HOST"
porta="PORT"
nome="DATABASE"
string=f"postgresql+psycopg://{usuario}:{senha}@{host}:{porta}/{nome}"
engine = create_engine(string)
print("Initiating")
for arquivo in os.listdir(caminho): #Find the folder
    if arquivo.endswith(".csv"):  #Find files .csv
        tabela = arquivo.replace(".csv","") #Change the file type
        final = os.path.join(caminho, arquivo) 
        print(f"Loading {tabela}") 
        try:
            df=pd.read_csv(final) #Read table
            tabela = tabela.lower() 
            df.drop(columns=df.columns[-2:],inplace=True) #Drop unnecessary columns
            df.to_sql(name=tabela, con=engine, if_exists="replace", index = False) 
            print(f"Sucess load {tabela}") #Table added to database
        except Exception as e:
            print(f"Error load {tabela}:")
            print(e)
```
